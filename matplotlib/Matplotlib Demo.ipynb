```python
df = pd.read_csv('data/data.csv', index_col='Responder_id')
df
```

```python
df['LanguagesWorkedWith'] = df['LanguagesWorkedWith'].str.split(';')
```

```python
df
```

```python
python_devs = df[df['LanguagesWorkedWith'].str.contains('Python')].count()
python_devs
```

```python
language_counter = Counter()

for row in df['LanguagesWorkedWith']:
    language_counter.update(row.split(';'))

print(language_counter.most_common(15))
```

```python
languages = []
popularity = []
```

```python
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages.reverse()
popularity.reverse()

print(languages)
print(popularity)
```

```python
plt.style.use('fivethirtyeight')

plt.title('Most Popular Programming Languages')

plt.xlabel('Number of people who use')

plt.barh(languages, popularity)
```
